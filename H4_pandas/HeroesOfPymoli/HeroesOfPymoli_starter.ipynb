```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load,encoding ="ISO-8859-1")
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#find an array of unique players
unique_players = purchase_data["SN"].unique()
#change unique player array to panda dataframe so I can apply numpy methods on it
unique_players_df = pd.DataFrame(unique_players)
#count total number of unique players
total_players = unique_players_df.count()
#put total number of players into a data frame
total_players_df = pd.DataFrame(total_players)
#rename column to "Total Players"
total_players_df= total_players_df.rename(columns = {0 : "Total Players"})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find an array of unique items
unique_items = purchase_data["Item ID"].unique()
#count total number of unique items
total_items = pd.DataFrame(unique_items).count()
#find average price of items
avg_price = purchase_data["Price"].mean()
#round to 2 decimals and add dollar sign
#count total number of purchases
total_purchases = purchase_data["Purchase ID"].count()
#sum revenues
total_revenue= purchase_data["Price"].sum()
total_revenue
#create a summary data frame
summary = pd.DataFrame({
    "Number of Unique Items":total_items,
     "Average Price":avg_price,
    "Number of Purchases":total_purchases,
    "Total Revenue":total_revenue})
#round average price and total revenue to 2 decimals and add dollar sign
summary["Average Price"] = summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:.2f}".format)

#display summary table
summary
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate the total number of players with different gender
gender_count = purchase_data["Gender"].value_counts()
#calculate the percentage of players with different gender
gender_count_percent = (gender_count/float(total_players))*100
#create a summary data frame for gender 
summary_gender = pd.DataFrame({
    "Percent of Players":gender_count_percent,
     "Total Count":gender_count})
#format percent of players to 2 decimals
summary_gender["Percent of Players"] = summary_gender["Percent of Players"].map("{:.2f}".format)
#display summary table
summary_gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group the total number of purchases by different gender
gender_data = purchase_data.groupby("Gender")
#calculate the total number of purchase made by different gender
total_purchase_count = gender_data["Purchase ID"].count()
#calculate the average purchase price
average_price = gender_data["Price"].mean()
#sum total purchase amount
total_price = gender_data["Price"].sum()
#normalized purchase price
price_norm = total_price/total_purchase_count

#create a summary data frame
purchasing_analysis_summary = pd.DataFrame({
    "Purchase Count":total_purchase_count,
     "Average Purchase Price":average_price,
    "Total Purchase Value":total_price,
    "Normalized Totals":price_norm})

#format Average Purchase Price, Total Purchase Value and Normalized Totals to 2 decimals and add dollar sign
purchasing_analysis_summary[ "Average Purchase Price"] = purchasing_analysis_summary["Average Purchase Price"].map("$ {:.2f}".format)
purchasing_analysis_summary[ "Total Purchase Value"] = purchasing_analysis_summary["Total Purchase Value"].map("$ {:.2f}".format)
purchasing_analysis_summary[ "Normalized Totals"] = purchasing_analysis_summary[ "Normalized Totals"].map("$ {:.2f}".format)
#display summary table
purchasing_analysis_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#slice the data and place it into bins
purchase_data["group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
#group data by age category
age_data = purchase_data.groupby("group")
#count total number of players from different age
total_player_byage=age_data["SN"].count()
#calculate percent of players
percent_player_byage=(total_player_byage/float(total_players))*100

#create a summary data frame
dempgraphic_summary = pd.DataFrame({
    "Percent of Players":percent_player_byage,
     "Total Count":total_player_byage
})

#format Percent of Players to 2 decimals
dempgraphic_summary[ "Percent of Players"] =dempgraphic_summary["Percent of Players"].map("{:.2f}".format)
#display summary table
dempgraphic_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#slice the data and place it into bins
purchase_data["group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
#group data by age category
age_data = purchase_data.groupby("group")
#count total number of purchases made by different age
total_numberpurchase_byage=age_data["Purchase ID"].count()
#calculate average purchase price
average_price_byage=age_data["Price"].mean()
#sum total purchase value by each age group
total_purchase_byage=age_data["Price"].sum()
#normalized purchase price
price_norm_byage = total_purchase_byage/total_numberpurchase_byage
#a summary data frame
purchase_byage_summary = pd.DataFrame({
    "Purchase Count":total_numberpurchase_byage,
     "Average Purchase Price":average_price_byage,
    "Total Purchase Value":total_purchase_byage,
    "Normalized Totals":price_norm_byage
})

#format Average Purchase Price,Total Purchase Value and Normalized Total to 2 decimals and add dollar sign
purchase_byage_summary[ "Average Purchase Price"] =purchase_byage_summary["Average Purchase Price"].map("$ {:.2f}".format)
purchase_byage_summary[ "Total Purchase Value"] =purchase_byage_summary["Total Purchase Value"].map("$ {:.2f}".format)
purchase_byage_summary[ "Normalized Totals"] =purchase_byage_summary["Normalized Totals"].map("$ {:.2f}".format)

#display summary table
purchase_byage_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group data by age category
spender_data = purchase_data.groupby("SN")
#count total number of purchases made by each spender
purchase_count=spender_data["Purchase ID"].count()
#calculate average purchase price per spender
average_price_spender=spender_data["Price"].mean()
#sum total purchase value by each spender
total_purchase_spender=spender_data["Price"].sum()

#a summary data frame
spender_summary = pd.DataFrame({
   "Purchase Count":purchase_count,
    "Average Purchase Price":average_price_spender,
    "Total Purchase Value":total_purchase_spender,
})

#sort toal purchase value colum in descending order
spender_summary = spender_summary.sort_values("Total Purchase Value", ascending=False)

#format Average Purchase Price,Total Purchase Value to 2 decimals and add dollar sign
spender_summary[ "Average Purchase Price"] =spender_summary["Average Purchase Price"].map("$ {:.2f}".format)
spender_summary[ "Total Purchase Value"] =spender_summary["Total Purchase Value"].map("$ {:.2f}".format)

#display summary table
spender_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name and Price columns
item_data= purchase_data[["Item ID", "Item Name", "Price"]]
#Group data by Item ID and Item Name
grouped_item = item_data.groupby(["Item ID", "Item Name"])
#calculate number of purchase of each item
purchase_count_items = grouped_item["Item ID"].count()
#calculate total purchase value
totalpurchase_items = grouped_item["Price"].sum()

#display item price
item_price=grouped_item["Price"]
item_price.max()

#a summary data frame
item_summary = pd.DataFrame({
   "Purchase Count":purchase_count_items,
   "Item Price":item_price.max(),
    "Total Purchase Value":totalpurchase_items,
})

#sort the purchase count column in descending order
item_summary= item_summary.sort_values("Purchase Count", ascending= False)
#format Average Purchase Price,Total Purchase Value to 2 decimals and add dollar sign
item_summary[ "Item Price"] =item_summary["Item Price"].map("$ {:.2f}".format)
item_summary[ "Total Purchase Value"] =item_summary["Total Purchase Value"].map("$ {:.2f}".format)
item_summary.head()
```
