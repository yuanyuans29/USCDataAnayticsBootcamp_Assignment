```python
import os
import pandas as pd
import csv
import datetime

# Module for reading CSV files
csvpath = os.path.join('data', 'tripdatasummer2017.csv')

# Method 2: Improved Reading using CSV module

with open(csvpath,newline='') as csvfile:

    # CSV reader specifies delimiter and variable that holds contents
    csvreader = csv.reader(csvfile, delimiter=',')

    # Read the header row first (skip this step if there is now header)
    csv_header = next(csvreader)

    # Read each row of data after the header
    q1 = []
    starthourlist = []

    for row in csvreader:   
        x = datetime.datetime.strptime(row[1],"%Y-%m-%d %H:%M")
        starthourlist.append(x.hour)
        q1.append(row)

df1 = pd.DataFrame(q1, columns=csv_header)
df1["starthour"] = starthourlist
df1.head()
```

```python
summerhours = df1["starthour"].value_counts()
```
