```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
#import requests module
import requests


# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#import requests module
import requests


base_url = "http://api.openweathermap.org/data/2.5/weather?"

print("Beginning Data Retrieval")
print("-----------------------------")

info_list = []

#loop over each city in cities list
for city in cities:
    #create api requests url
    city_url = f"{base_url}q={city}&appid={api_key}&units=Imperial"
    #send api requests for each city
    response = requests.get(city_url)
    #sotre api requests response in a list
    city_dict = response.json()
    info_list.append(city_dict)
    
    #print log of each city as it's being processed with the city name
    if city_dict["cod"] == 200:
        print(f"Processing Record 1 of Set 1 | {city}  {city_url}")
    else:
        print(f"City not found. Skipping...")
print("-----------------------------")
print("Data Retrieval Complete")      
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#only includes cities with current weather data
info_list = [dict for dict in info_list if dict["cod"] == 200]

#create separate lists to store different data of each city
city = []
cloudiness=[]
country=[]
date =[]
humidity =[]
lat =[]
lon =[]
temperature = []
windspeed = []

#loop over json response to extract city name, cloudiness, country code, 
#latitude, longtitude, max temperature and wind speed of each city

for element in info_list:
    n = element["name"]
    c = element["clouds"]["all"]
    cy = element["sys"]["country"]
    d = element["dt"]
    h = element["main"]["humidity"]
    la = element["coord"]["lat"]
    ln = element["coord"]["lon"]
    t = element["main"]["temp_max"]
    s = element["wind"]["speed"]

#add data extracted from json response to new lists created above
    city.append(n)
    cloudiness.append(c)
    country.append(cy)
    date.append(d)
    humidity.append(h)
    lat.append(la)
    lon.append(ln)
    temperature.append(t)
    windspeed.append(s)

#create a dataframe with all extracted data
weather_list = [city,cloudiness,country, date,humidity,lat,lon,temperature,windspeed]
weather_df = pd.DataFrame(weather_list)
weather_df = weather_df.transpose()
weather_df.columns = ["city","cloudiness","country", "date", "humidity","lat","lon","temperature","windspeed"]

#count
weather_df.count()
```

```python
#explort cities data to a csv file
weather_df.to_csv("../output_data/cities.csv",index=True)

#print out weather dataframe
weather_df.head()
```
